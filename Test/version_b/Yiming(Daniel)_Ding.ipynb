#### 8 - (12 marks)
Write a function with the following specifications:

|               |                           |
|:--------------|:--------------------------|
| Name:         | `is_multiple_of_six`      |
| Arguments:    | A `number`                |
| Returns:      | `Bool`                    |
| Displays:     | Nothing                   |
| Tasks:        | Returns `True` if the argument is a multiple of 6 otherwise returns `False`   |
|               |                           |
| Test Harness: | Call the function three times with 11, 12 and 13. You **MUST** match the output exactly    |

````python
def is_multiple_of_six(number):
    if number % 6 == 0:
        print(f'is {number} divisible by 6? => True')
        return True
        
    else:
        print(f'is {number} divisible by 6? => False')
        return False
is_multiple_of_six(11)
is_multiple_of_six(12)
is_multiple_of_six(13)
````

#### 9 - (12 marks)
Write a function with the following specifications:

|               |                           |
|:--------------|:--------------------------|
| Name:         | `area_of_cylinder`        |
| Arguments:    | A `height` and a `radius` |
| Returns:      | A float                   |
| Displays:     | Nothing                   |
| Tasks:        | Calculates and returns the surface area of a cylinder using the radius and height specified by the arguments  |
|               |                           |
| Test Harness: | Call the function three times with appropriate arguments to produce the required output |

*(Hint: area of cylinder = 2 * 3.14 * radius^2 * height)*

````python
def area_of_cylinder(height, radius):
    area = 2 * 3.14 * radius**2 * height
    print(f'A cylinder of height {height:.1f} and radius of {radius:.1f} will have a surface area of {area:.2f}')
    return area
area_of_cylinder(3,2)
area_of_cylinder(4,2)
area_of_cylinder(4,10)
````

#### 10 - (19 marks)
Write a function with the following specifications:

|               |                           |
|:--------------|:--------------------------|
| Name:         | `print_appreciation_table`  |
| Arguments:    | `years`, `start_value` and `appreciation`  |
| Returns:      | Nothing                   |
| Displays:     | The output as shown by the sample    |
| Tasks:        | The third column is obtained by adding the appreciation to the end of year value. The fourth column is obtained by adding the appreciation to the existing accumulated appreciation|
|               |                          |
| Test Harness: | Call the function with appropriate arguments to produce the required output |

*(Hint: The function will have a loop but the table header should not be in the loop)*


#### Test harness:
```python
year, start_value, appreciation = 7, 10_000, 1_500
print_appreciation_table(year, start_value, appreciation)
```

````python
def print_appreciation_table(years, start_value, appreciation):
    print('                            END-OF-YEAR        ACCUMULATED')
    print('YEAR      DEPRECIATION        VALUE           DEPRECIATION')
    print('----      ------------      -----------       ------------')
    counter = 1
    total_de = 0
    second = f'${appreciation:,.2f}'
    while counter <= years:
        third = f'${start_value:,.2f}'
        fourth = f'${total_de:,.2f}'
        print(f'{counter:>3}{second:>19}{third:>17}{fourth:>19}')
        total_de = total_de + appreciation
        start_value = start_value - appreciation
        counter += 1

print_appreciation_table(7, 25000, 3500)
````
