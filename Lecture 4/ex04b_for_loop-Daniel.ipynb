#### 19
Write a program to find al 3-digits Armstrong number (e.g. 153 = 1^3 + 5^3 + 3^3)

```python
#Looping through 100 to 999
for num in range(100, 1000, 1):
    #making int into string
    char_num = str(num)
    #Declaring total = 0 at the first loop since we want to clear it after we 
    #have done with comparing total and num.
    total = 0
    #Creating a loop within a loop so we can loop though each character of str(num)
    for char in char_num:
        #Each digit is retrieved by looking at each character of
        #str(num) and making it into a integer
        digit = int(char)
        #Adding all the digit to see if num is a armstrong number
        total += digit ** 3
        #test if num is equal to total, if yes print it
        if total == num:
            #remember if u want the first character of the strong it starts from 0.
            print(f'{num} = {char_num[0]}^3 + {char_num[1]}^3 + {char_num[2]}^3')
        
```

```python
print("3-digit Armstrong numbers are:")
#Loop should be inside 100 and 999.
for num in range(100, 1000):
    #This is list comprehension
    digits = [int(d) for d in str(num)]
    if sum(d**3 for d in digits) == num:
        print(f"{num} = {' + '.join(f'{d}³' for d in digits)}")


```
