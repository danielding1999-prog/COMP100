### 1	

Write a program to print the numbers 2 to 10 in increments of two on the same line. The output of your program should be 2 4 6 8 10 

```python
count = 2
end = 10
while count <= end:
    print(count)
    count += 2
```

#### 7 
Write a program to print all the numbers from 1 tp 50 that are multiples of 3 or 5 or both.

```python
count = 1
end = 51
while count < end:
    if count % 3 == 0 or count % 5 == 0:
        print(count)
    count += 1
```

#### 8
Write a program to print all the numbers from 40 to 0 that are NOT multiples of 7.

```python
count = 40
end = 0
while count >= end:
    if count % 7 != 0:
        print(count)
    count -= 1
```

Write a python program that displays a conversion table from gallons to liters. The program should display gallons from 10 to 20 in one-gallon increments and the corresponding liter equivalents. Use the relationship that 1 gallon contains 3.785 liters. Your table must look professional!

```python
#from 10 to 20 is 11 entries
header = 'Gallon      Litters'
litter = 0
print(header)
print('=' * len(header))
for gallon in range(10, 21, 1):
# range function takes in first range and second and last entry is the increment
   
    litter = f'{(gallon * 3.785):.2f}'
    print(f'{gallon:<10}{litter:>7}')
```

Thus, instead of the condition checking for a fixed count, the condition checks for the ending Fahrenheit value. Your table must look professional!

Your output should be formatted as shown below
|meters |  feet|
| ---: | ---: |
|    0 |      0.00 |
|    3 |      9.75 |
|    6 |     19.50 |
|    9 |     29.25 |
|   12 |     39.00 |

|   18 |     58.50 |
|   21 |     68.25 |
|   24 |     78.00 |
|   27 |     87.75 |
|   30 |     97.50 |

```python
start = float(input('Please enter the starting Fahrenheit.'))
end = float(input('Please enter the last Fahrenheit.'))
increment = int(input('Please enter the increment.'))

for feh in range(start, (end + increment), increment):
    print(feh)
```
