#### 9
Decomposition (Unpacking)

Given:

student = ["Diana", 88, 91, 95]


-   Unpack so that name = "Diana" and grades = [88, 91, 95]
-   Unpack into name, g1, g2, g3

```python
student = ["Diana", 88, 91, 95]
name, *grades = student
print(name)
print(grades)
g1, g2, g3 = grades
print(f'{g1} {g2} {g3}')
```
