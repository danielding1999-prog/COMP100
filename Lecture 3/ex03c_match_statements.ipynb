## Simple Matching

Write a program that asks the user to enter "red", "green", or "blue".
Use match to print a message like "You chose red!", "You chose green!", or "You chose blue!".
If the input doesn’t match, print "Unknown color".


```python
color = input('Please enter color red, green or blue')
match (color.lower()):
    case 'red':
        print(f'You choose {color.lower()}')
    case 'blue':
        print(f'You choose {color.lower()}')
    case 'green':
        print(f'you choose {color.lower()}')
    case _:
        print('Unknown color')
```

(Hint: lowercase the input first with .lower().)

```python
day = input('Please enter a day of the week.')
match (day.lower()):
    case 'sunday' | 'saturday' | '6' | '7':
        print("Weekend")
    case 'monday' | 'tuesday' | 'wednesday' | 'thursday' | 'friday'| '1' | '2' | '3' | '4' | '5':
        print('Weekday')
    case _:
        print('Unknown day')
```

## Guards (with if)

Write a program that asks the user for an integer.
Use match to check:

If the number is negative

If the number is even

Otherwise, treat it as odd

```python
integer = int(input('Please enter a integer.'))
match integer:
    case x if x < 0:
        print('Negative')
    case x if x > 0:
        print('Positive')
    case _:
        print('Odd')
```

## Combining Patterns

Write a program that asks the user to enter a number between 1 and 10.
Use match to:

Print "One" if the number is 1

Print "Two or Three" if the number is 2 or 3

Print "Greater than 5" if the number is > 5

Print "Other number" otherwise

```python
num = float(input('Please enter a number between 1 and 10.'))
match num:
    case 1:
        print('One.')
    case 2 | 3:
        print('Two or Three.')
    case x if x > 5:
        print('Greater than 5.')
    case _:
        print('Other number.')
```

## Matching Sequences

Given a list of names, use match to:

Print "Two names: ..." if the list has exactly two items

Print "Three names: ..." if the list has exactly three items

Print "Many names: ..." if the list has more than three items

Test with:

["Alice", "Bob"]
["Alice", "Bob", "Charlie"]
["Alice", "Bob", "Charlie", "David"]



```python
name = input('Please enter name with space in between')
name_list = name.split()
match name_list:
    case [x , y]:
        print(f'Two names: {x}, {y}')
    case [x , y , z]:
        print(f'Three names: {x}, {y}, {z}')
    case [x, *y , z]:
        print(f'Many names: {x}, {', '.join(y)}, {z}')
    case _:
        print("Invalid")
```
