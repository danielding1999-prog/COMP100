### Using Built-in Functions

#### type() function

```python
# Use type() to check the type of each of the following variables:
# 
x = 42;
y = 3.14;
z = "Python";
is_active = True;
print(type(x));
print(type(y));
print(type(z));
print(type(is_active));
```

#### id() function

```python
# Use id() to check whether two variables pointing to the same value actually point to the same object:
# 
a = 100;
b = 100;
print(id(a));
print(id(b));
# It is pointed to the same object since a and b point to the same object.
```

#### dir() function

```python
# Part 4 – Input and Eval
# 
# Write a small program that:
# 
# Asks the user for their name
name = input('What is your name?')
# Asks the user for their age
age = int(input('What is your age?'))
# Prints a greeting like: "Hello Alice, you are 20 years old."
print('Hello',name,', you are', age,"years old.")
```

#### eval()

```python
# Use eval() to evaluate the string "2 + 3 * 4".
print(eval("2 + 3 * 4"))
# Compare the result with what you expected.
a = 2 + 3 * 4
print(a)
print(a+1)
# it actually evaluated a string.

# Bonus discussion: Why should we avoid eval() in real-world programs?
# Security risk, since eval function allows python program to run python within it self.
```
